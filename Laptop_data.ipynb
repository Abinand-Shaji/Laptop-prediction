**Created highcorelated dataframe**

```python
target=abs(laptop_data_org.corr()['Price'])
features=target[target>0.1].index
corelated_data=laptop_data_org[features]
corelated_data.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(corelated_data.corr(),annot=True,cmap='magma')
```

**Model creation and prediction Using pipeline**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error, r2_score,mean_absolute_error

# Splitting input and output
input_features = corelated_data.drop(columns=['Price'], axis=1)
output_target = corelated_data['Price']

# Splitting data into training and testing sets
input_train, input_test, output_train, output_test = train_test_split(
    input_features, output_target, test_size=0.20, random_state=1)

# Define models and their names
models = [
    ('Linear Regression', LinearRegression()),
    ('DecisionTree',DecisionTreeRegressor() ),
    ('RandomForest',RandomForestRegressor())
]

# Iterate through models, create pipelines, and evaluate
for model_name, model in models:
    pipe = Pipeline([(model_name, model)])



    pipe.fit(input_train, output_train)
    predictions = pipe.predict(input_test)

    mape=mean_absolute_percentage_error(output_test,predictions)
    r2 = r2_score(output_test, predictions)
    mae=mean_absolute_error(output_test,predictions)


    print(f"--- {model_name} ---")
    print("mean_absolute_error",mae)
    print('Mean_absolute_percentage_error',mape)
    print("R2 score", r2)
    print()
```
